```python
# Read the contents of the jfng360k.pos file to understand its structure
with open("./jfng360k.pos", "r") as file:
    jfng360k_lines = file.readlines()

# Display the first few lines to understand the structure
jfng360k_lines[:10]
```

```python
# Find the start of the actual data by looking for lines that don't start with '%'
data_start_index_k = next(i for i, line in enumerate(jfng360k_lines) if not line.startswith('%'))

# Display a few lines starting from the data to understand its structure
jfng360k_lines[data_start_index_k:data_start_index_k+10]
```

```python
import pandas as pd

# Extract data and convert to a DataFrame
columns = jfng360k_lines[data_start_index_k].split(',')
data_k = [line.split(',') for line in jfng360k_lines[data_start_index_k+1:]]
df_k = pd.DataFrame(data_k, columns=columns)

# Strip whitespace from column names to fix the KeyError
df_k.columns = [col.strip() for col in df_k.columns]

# Convert relevant columns to float
df_k['e-baseline(m)'] = df_k['e-baseline(m)'].astype(float)
df_k['n-baseline(m)'] = df_k['n-baseline(m)'].astype(float)
df_k['u-baseline(m)'] = df_k['u-baseline(m)'].astype(float)

# Calculate velocities using the diff() method
df_k['ve(m/s)'] = df_k['e-baseline(m)'].diff()  # Velocity in East direction
df_k['vn(m/s)'] = df_k['n-baseline(m)'].diff()  # Velocity in North direction
df_k['vu(m/s)'] = df_k['u-baseline(m)'].diff()  # Velocity in Up direction

# Drop the first row since it doesn't have velocity information
df_k = df_k.dropna(subset=['ve(m/s)', 'vn(m/s)', 'vu(m/s)'])

# Extract the calculated velocities
velocities_k = df_k[['GPST', 've(m/s)', 'vn(m/s)', 'vu(m/s)']]

# Display the first few rows of the calculated velocities
velocities_k.head()

```

```python
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

# Filter out data with velocities greater than 2cm/s (or 20mm/s)
df_k_filtered = df_k[(df_k['ve(m/s)'].abs() <= 0.1) & 
                     (df_k['vn(m/s)'].abs() <= 0.1) & 
                     (df_k['vu(m/s)'].abs() <= 0.1)]

# Convert velocities to mm/s
df_k_filtered['ve(m/s)'] *= 1000
df_k_filtered['vn(m/s)'] *= 1000
df_k_filtered['vu(m/s)'] *= 1000

# # Convert velocities to mm/s
# df_k['ve(m/s)'] *= 1000
# df_k['vn(m/s)'] *= 1000
# df_k['vu(m/s)'] *= 1000

# Plotting the velocities over time
fig, ax = plt.subplots(3, 1, figsize=(14, 12), sharex=True)

# East (E) velocity
ax[0].plot(pd.to_datetime(df_k_filtered['GPST']), df_k_filtered['ve(m/s)'], label='East', color='r')
ax[0].set_ylabel('Velocity (mm/s)')
ax[0].set_title('East Velocity over Time')
ax[0].grid(True)

# North (N) velocity
ax[1].plot(pd.to_datetime(df_k_filtered['GPST']), df_k_filtered['vn(m/s)'], label='North', color='g')
ax[1].set_ylabel('Velocity (mm/s)')
ax[1].set_title('North Velocity over Time')
ax[1].grid(True)

# Up (U) velocity
ax[2].plot(pd.to_datetime(df_k_filtered['GPST']), df_k_filtered['vu(m/s)'], label='Up', color='b')
ax[2].set_ylabel('Velocity (mm/s)')
ax[2].set_title('Up Velocity over Time')
ax[2].grid(True)
ax[2].set_xlabel('Time')

date_format = DateFormatter("%H:%M")
ax[2].xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.show()
```

```python
# Plotting histograms for East, North, and Up velocities
fig, ax = plt.subplots(3, 1, figsize=(5, 12))

# East (E) velocity histogram
ax[0].hist(df_k_filtered['ve(m/s)'], bins=100, color='r', edgecolor='black')
ax[0].set_ylabel('Frequency')
ax[0].set_title('Histogram of East Velocity')
ax[0].grid(True, which='both', linestyle='--', linewidth=0.5)

# North (N) velocity histogram
ax[1].hist(df_k_filtered['vn(m/s)'], bins=100, color='g', edgecolor='black')
ax[1].set_ylabel('Frequency')
ax[1].set_title('Histogram of North Velocity')
ax[1].grid(True, which='both', linestyle='--', linewidth=0.5)

# Up (U) velocity histogram
ax[2].hist(df_k_filtered['vu(m/s)'], bins=100, color='b', edgecolor='black')
ax[2].set_xlabel('Velocity (mm/s)')
ax[2].set_ylabel('Frequency')
ax[2].set_title('Histogram of Up Velocity')
ax[2].grid(True, which='both', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```
