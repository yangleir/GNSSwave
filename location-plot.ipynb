# 绘制基站位置示意图

```python
from matplotlib.path import Path
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.io.img_tiles import OSM, MapboxTiles, Stamen, GoogleTiles
```

```python
jfng = 30.515567412, 114.491023090
wuh2 = 30.531679047, 114.357268134
```

```python
imagery = OSM()

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot(1, 1, 1, projection=imagery.crs)
ax.set_extent([114.34, 114.515, 30.48, 30.57], ccrs.PlateCarree())

# Add the imagery to the map.
ax.add_image(imagery, 16)

ax.plot([wuh2[1], jfng[1]], [wuh2[0], jfng[0]], color='dimgray', transform=ccrs.PlateCarree())
ax.plot(jfng[1], jfng[0], transform=ccrs.Geodetic(), marker='o')
ax.plot(wuh2[1], wuh2[0], color='darkred', transform=ccrs.Geodetic(), marker='*')
ax.text(jfng[1]-0.007, jfng[0]-0.006, 'JFNG', transform=ccrs.Geodetic(), fontfamily='serif')
ax.text(wuh2[1]-0.008, wuh2[0]-0.006, 'WUH2', transform=ccrs.Geodetic(), fontfamily='serif')
ax.text(114.373, 30.553, 'Dong Lake', transform=ccrs.Geodetic(), fontfamily='serif', size='x-small')
ax.text(114.347, 30.49, 'Nan Lake', transform=ccrs.Geodetic(), fontfamily='serif', size='x-small')
ax.text(114.411, 30.492, 'Wuhan', transform=ccrs.Geodetic(), fontfamily='serif', size='x-large')
ax.text(114.422, 30.524, '13km', transform=ccrs.Geodetic(), fontfamily='serif', rotation=-8.5, size='small')
ax.gridlines(ls='-.', draw_labels=["bottom", "left", "right"], dms=True, x_inline=False, rotate_labels=False)

# Plot global map as a subfigure
sub_ax = fig.add_axes([0.63, 0.557, 0.32, 0.32],
                      projection=ccrs.Orthographic(central_latitude=27, central_longitude=108))
# sub_ax.stock_img()
sub_ax.add_feature(cfeature.LAND, facecolor='darkslategrey')
sub_ax.plot(114.4, 30.5, 'o', color='firebrick', transform=ccrs.Geodetic())
# Plot gridlines
sub_ax.gridlines(ls='-.')
sub_ax.set_global()

plt.savefig('../assets/igs-location.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
